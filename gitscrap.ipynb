```python
import requests

def fetch_contents(url, path='', depth=0):
    """
    Fonction récursive pour récupérer et afficher le contenu des dossiers GitHub sous forme d'arbre.
    :param url: URL de base de l'API pour le contenu du dépôt.
    :param path: Chemin relatif à explorer dans le dépôt.
    :param depth: Profondeur actuelle dans l'arbre de répertoire.
    """
    headers = {"Authorization": f"token {GITHUB_TOKEN}"}
    full_url = f"{url}{path}" if path else url  # Construire l'URL complète
    response = requests.get(full_url, headers=headers)
    
    if response.status_code == 200:
        contents = response.json()
        indent = '    ' * depth  # Créer l'indentation en fonction de la profondeur
        for item in contents:
            print(f"{indent}- {item['name']} ({item['type']})")
            if item['type'] == 'dir':  # Si l'élément est un dossier, explorez-le récursivement
                fetch_contents(url, item['path'], depth + 1)
    else:
        print("Erreur lors de la récupération du contenu:", response.status_code)

# Configuration initiale
GITHUB_TOKEN = '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'
owner = 'donnemartin'
repo = 'system-design-primer'

# URL pour accéder au contenu de la racine du dépôt
url_contents = f'https://api.github.com/repos/{owner}/{repo}/contents/'

# Démarrer l'exploration à partir de la racine du dépôt
fetch_contents(url_contents)
```
